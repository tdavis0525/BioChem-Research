```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


twopi= int( 3.14159 * 2.0)

# Define the range for x and y
x = np.linspace(-twopi, twopi, twopi)
y = np.linspace(-5, 5, 50)

# Create a meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Calculate the z values: sum of two sine waves, normalized to 0-1 range
Z = np.sin(X) + 0.5 * np.sin(2*Y)
Z_normalized = (Z - Z.min()) / (Z.max() - Z.min())  # Normalize to 0-1

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z_normalized) # Use the normalized Z values

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Probability')  # Changed z-label to "Probability"
ax.set_title('3D Probability Range') # Changed Title
plt.show()
```
